## Real GDP

```python
df_real_gdp = pd.read_excel('GDP real (2000 onwards).xlsx')
print("DataFrame 'df_real_gdp' loaded successfully.")
df_real_gdp
```

```python
df_real_gdp = df_real_gdp.iloc[[3, 4], 1:]
print("DataFrame 'df_real_gdp' modified to include only rows 3 and 4, and exclude the first column.")
df_real_gdp
```

```python
# Transpose rows and columns
gdp_df = df_real_gdp.T
# Rename the columns from the original row indices (3 and 4) to 'Date' and 'Real GDP'
gdp_df.rename(columns={3: 'Date', 4: 'Real GDP'}, inplace=True)

# Convert 'Date' column to datetime objects
gdp_df['Date'] = pd.to_datetime(gdp_df['Date'])
# Convert 'Real GDP' column to numeric, coercing errors to NaN
gdp_df['Real GDP'] = pd.to_numeric(gdp_df['Real GDP'], errors='coerce')
# Drop rows with NaN values in 'Real GDP' column that might have resulted from conversion errors
gdp_df = gdp_df.dropna(subset=['Real GDP'])

# Reset index, as the old column names (e.g., 'FactSet', 'Unnamed: 2') are no longer relevant as an index
gdp_df = gdp_df.reset_index(drop=True)

print("DataFrame 'gdp_df' created successfully with 'Date' and 'Real GDP' columns.")
```

```python
gdp_df
```

```python
gdp_df['Date'] = (gdp_df['Date'].dt.to_period('M') + 1).dt.to_timestamp()

print("Dates in 'gdp_df' updated to the first day of the next month.") #justification is from comparing CPI values of CEIC and Factset
gdp_df.head()
gdp_df.tail()
```

# FX Rate

```python
df_fx_rate = pd.read_excel('pesodollar.xlsx', header=None)

# Data starts from the 7th row (index 6).
# - 2nd column (index 1) contains the year
# - 3rd column (index 2) contains the month
# - 4th column (index 3) contains the FX rate values
df_working = df_fx_rate.iloc[6:, [1, 2, 3]].copy()
df_working.columns = ['Year', 'Month', 'FX Rate']

# The 'Year' column has NaNs that need to be forward-filled.
df_working['Year'] = df_working['Year'].ffill()

# Drop rows where 'Month' or 'FX_Rate' are NaN, as these are not complete data entries.
df_working = df_working.dropna(subset=['Month', 'FX Rate']).copy()

# Convert month names to numerical representation.
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}
df_working['Month_Num'] = df_working['Month'].map(month_mapping)

# Ensure 'Year' is integer type for date creation.
df_working['Year'] = df_working['Year'].astype(int)

# Create the 'Date' column combining 'Year', 'Month_Num', and setting the day to 1.
# Rename columns to 'year' and 'month' as pd.to_datetime expects these lowercase names
# when passed a DataFrame with date components.
df_working['Date'] = pd.to_datetime(
    df_working[['Year', 'Month_Num']].assign(day=1).rename(
        columns={'Year': 'year', 'Month_Num': 'month'}
    )
)

# Select the final desired columns: 'Date' as the 1st and 'FX_Rate' as the 2nd.
df_fx_rate_cleaned = df_working[['Date', 'FX Rate']].copy()

print("Cleaned and processed DataFrame 'df_fx_rate_cleaned'")

print("\nDataFrame Info:")
df_fx_rate_cleaned.info()
```

# transform Yoy

```python
main_df.info()
```

```python
cols_to_transform = [
    "Real GDP", "CPI", "Wholesale Price", "Industrial Production",
    "Intl Trade Merch Exports", "Intl Trade Merch Imports", "FX Rate"
]

for c in cols_to_transform:
    # Check if strictly positive (needed for logs)
    if (main_df[c] > 0).all():
        main_df[f"{c}_yoy_log"] = 100 * (np.log(main_df[c]) - np.log(main_df[c].shift(12)))
    else:
        main_df[f"{c}_yoy_pct"] = 100 * (main_df[c] / main_df[c].shift(12) - 1)
```
